# Transform Image to PPM Format

```python
import cv2
im_rgb = cv2.imread("SPPM/objects/desert.jpg")
# default load BGR
h, w, _ = im_rgb.shape
print(h, w)
```

```python
# PPM requires RGB
with open("SPPM/objects/desert.ppm", "w") as f:
    f.write("P3\n"+str(w)+" "+str(h)+"\n255\n")
    for j in range(h):
        for i in range(w):
            f.write(str(im_rgb[j][i][2])+" "+str(im_rgb[j][i][1])+" "+str(im_rgb[j][i][0])+" ")
```

# Change Obj file to Ply file

```python
nv = 0
nm = 0
with open("SPPM/objects/relic.obj", "r") as f:
    for i in f.readlines():
        if(i[0]=="v"):
            if(nm!=0):
                print(i)
            nv += 1
        elif(i[0]=="f"):
            print(i)
            nm += 1
        else:
            print("ERROR")
```
